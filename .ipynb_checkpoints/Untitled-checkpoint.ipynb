```python
#normalisasi data
temp={'suhu':[],'zt_tds':[],'zt_tss':[],'ph':[],'cod':[],'nitrat':[],'minyak':[],'bk':[],'bkt':[]}
for dt in data_training:
        
        temp['suhu'].append(float(dt['suhu']))
        temp['zt_tds'].append(float(dt['zt_tds']))
        temp['zt_tss'].append(float(dt['zt_tss']))
        temp['ph'].append(float(dt['ph']))
        temp['cod'].append(float(dt['cod']))
        temp['nitrat'].append(float(dt['nitrat']))
        temp['minyak'].append(float(dt['minyak']))
        temp['bk'].append(float(dt['bk']))
        temp['bkt'].append(float(dt['bkt']))

        max_suhu = max(temp['suhu'])
        max_zt_tds = max(temp['zt_tds'])
        max_zt_tss = max(temp['zt_tss'])
        max_ph = max(temp['ph'])
        max_cod = max(temp['cod'])
        max_nitrat = max(temp['nitrat'])
        max_minyak = max(temp['minyak'])
        max_bk = max(temp['bk'])
        max_bkt = max(temp['bkt'])
        
data={'suhu':[],'zt_tds':[],'zt_tss':[],'ph':[],'cod':[],'nitrat':[],'minyak':[],'bk':[],'bkt':[]}
for i in temp['suhu']:
        data['suhu'].append(i/max_suhu)
for j in temp['zt_tds']:
        data['zt_tds'].append(j/max_zt_tds)
for k in temp['zt_tss']:
        data['zt_tss'].append(k/max_zt_tss)
for l in temp['ph']:
        data['ph'].append(l/max_ph)
for m in temp['cod']:
        data['cod'].append(m/max_cod)
for n in temp['nitrat']:
        data['nitrat'].append(n/max_nitrat)
for r in temp['minyak']:
        data['minyak'].append(r/max_minyak)
for s in temp['bk']:
        data['bk'].append(s/max_bk)
for p in temp['bkt']:
        data['bkt'].append(p/max_bkt)
```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```
