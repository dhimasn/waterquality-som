```python
#Nama : Dhimas Nandsita
#NIm : G6501211010

import mysql.connector
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt

class SOM :
    # Function here computes the winning vector
    # by Euclidean distance
    def winner( self, weights, sample ) :
        
        D = {'0':0,'1':0,'2':0}
        for i in range( len( sample ) ) :
            D['0'] = D['0'] + math.pow( ( sample[i] - weights[0][i] ), 2 )
            D['1'] = D['1'] + math.pow( ( sample[i] - weights[1][i] ), 2 )
            D['2'] = D['2'] + math.pow( ( sample[i] - weights[2][i] ), 2 )

            d = max(D['0'], D['1'], D['2'])
            if(d == D['0']):
                return 0
            if(d == D['1']):
                return 1
            if(d == D['2']):
                return 2
            
    # Function to testing
    def winnerTest ( self, weights, sample) :
        result ={'score':[],'kelas':[],'id':[]}
        
        for i in range( len( sample ) ):
            
            D = {'0':0,'1':0,'2':0}
            
            for j in range (len(sample.loc[0])):
            
                D['0'] = D['0'] + math.pow( ( sample.loc[i][j] - weights[0][j] ), 2 )
                D['1'] = D['1'] + math.pow( ( sample.loc[i][j] - weights[1][j] ), 2 )
                D['2'] = D['2'] + math.pow( ( sample.loc[i][j] - weights[2][j] ), 2 )

            d = max(D['0'], D['1'], D['2'])
            if(d == D['0']):
                result['id'].append(i)
                result['score'].append(D['0'])
                result['kelas'].append('Cemar Berat')
            if(d == D['1']):
                result['id'].append(i)
                result['score'].append(D['1'])
                result['kelas'].append('Cemar Sedang')
            if(d == D['2']):
                result['id'].append(i)
                result['score'].append(D['2'])
                result['kelas'].append('Cemar Ringan')
                    
        return result

    # Function here updates the winning vector
    def update( self, weights, sample, J, alpha ) :

        for i in range(len(weights[0])) :
            weights[J][i] = weights[J][i] + alpha * ( sample[i] - weights[J][i] )

        return weights
```

```python
#normalisasi data
temp={'suhu':[],'zt_tds':[],'zt_tss':[],'ph':[],'cod':[],'nitrat':[],'minyak':[],'bk':[],'bkt':[]}
for dt in data_training:
        
        temp['suhu'].append(float(dt['suhu']))
        temp['zt_tds'].append(float(dt['zt_tds']))
        temp['zt_tss'].append(float(dt['zt_tss']))
        temp['ph'].append(float(dt['ph']))
        temp['cod'].append(float(dt['cod']))
        temp['nitrat'].append(float(dt['nitrat']))
        temp['minyak'].append(float(dt['minyak']))
        temp['bk'].append(float(dt['bk']))
        temp['bkt'].append(float(dt['bkt']))

        max_suhu = max(temp['suhu'])
        max_zt_tds = max(temp['zt_tds'])
        max_zt_tss = max(temp['zt_tss'])
        max_ph = max(temp['ph'])
        max_cod = max(temp['cod'])
        max_nitrat = max(temp['nitrat'])
        max_minyak = max(temp['minyak'])
        max_bk = max(temp['bk'])
        max_bkt = max(temp['bkt'])
        
data={'suhu':[],'zt_tds':[],'zt_tss':[],'ph':[],'cod':[],'nitrat':[],'minyak':[],'bk':[],'bkt':[]}
for i in temp['suhu']:
        data['suhu'].append(i/max_suhu)
for j in temp['zt_tds']:
        data['zt_tds'].append(j/max_zt_tds)
for k in temp['zt_tss']:
        data['zt_tss'].append(k/max_zt_tss)
for l in temp['ph']:
        data['ph'].append(l/max_ph)
for m in temp['cod']:
        data['cod'].append(m/max_cod)
for n in temp['nitrat']:
        data['nitrat'].append(n/max_nitrat)
for r in temp['minyak']:
        data['minyak'].append(r/max_minyak)
for s in temp['bk']:
        data['bk'].append(s/max_bk)
for p in temp['bkt']:
        data['bkt'].append(p/max_bkt)
```
